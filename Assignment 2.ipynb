```python
# Q4
def prime_fact(x):
    p_factors = []
    div = 2
    
    while div <= x:
        if x%div == 0:
            p_factors.append(div)
            x = x/div
        else:
            div += 1
    return p_factors

a = int(input("Enter a number :"))
print(prime_fact(a))
```

```python
# Q5
list1 = []
 
def second_largest(l):
    new_list = set(l)
    new_list.remove(max(new_list))
    return max(new_list)

for i in range(1, 5):
    ele = int(input("Enter elements: "))
    list1.append(ele)
    
print('Original elements: ',list1)
print('2nd largest among 4 numbers: ',second_largest(list1))
```

```python
# Q6
s = "hello I am good hello user user welcome"
l = s.split()
k = []
for i in l:

	# If condition is used to store unique string
	# in another list 'k'
	if (s.count(i)>=1 and (i not in k)):
		k.append(i)
print(' '.join(k))
```

```python
# Q8
print("Enter Two Numbers: ", end="")
no = int(input())
nt = int(input())

a = no
b = nt
while b!=0:
    temp = b
    b = a%b
    a = temp

hcf = a
lcm = int((no*nt)/hcf)

print("\nLCM (" + str(no) + ", " + str(nt) + ") = ", lcm)
print("HCF (" + str(no) + ", " + str(nt) + ") = ", hcf)
```
